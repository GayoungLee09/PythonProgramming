```python
#2.끝말잇기
import random
f = open('dict_test.txt','r')
voc = []
v=[]
vocab = []
wordlist = []
for i in f:
    voc = i.strip().split(' :')
    if 'n.' in voc[1]:
        v.append(voc[0])
for j in v:
    if len(j)==5:
        vocab.append(j)
start = random.choice(vocab)
wordlist.append(start)
    
while True:
    word = input("%s 끝말 잇기? "%wordlist[-1])
    if word in wordlist:
        print("중복된 단어입니다(%s의 끝말을 이으세요)."%wordlist[-1])
    else:
        if len(word) != 5:
            if len(word) > 5:
                print("단어가 길어요(%s의 끝말을 이으세요)."%wordlist[-1])
            else:
                print("단어가 짧아요(%s의 끝말을 이으세요)."%wordlist[-1])
        else:
            if word in vocab:
                wordlist.append(word)
                print("정답입니다(%s의 끝말을 이으세요)." %wordlist[-1])
            else:
                print("사전에 없는 단어입니다(%s의 끝말을 이으세요)."%wordlist[-1])
    
f.close()
```
